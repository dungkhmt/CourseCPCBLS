```python
import json
import os
from ortools.sat.python import cp_model
import csv

coeff = 1e6

data_path = 'src/khmtk60/miniprojects/multiknapsackminmaxtypeconstraints/MinMaxTypeMultiKnapsackInput-1.json'

ori_data = json.load(open(data_path))

data = ori_data
items,bins = data['items'],data['bins']
n, m = len(items), len(bins)
ncat = max([item['t'] for item in items])+1
ncla = max([item['r'] for item in items])+1

for item in items:
    item['w'], item['p'] = int(item['w']*coeff), int(item['p']*coeff)

for b in bins:
    b['capacity'], b['minLoad'], b['p'] = int(b['capacity']*coeff), int(b['minLoad']*coeff), int(b['p']*coeff)
```

```python
model = cp_model.CpModel()

x = []
for i in list(range(n)):
    x.append([model.NewBoolVar(f'x_{j}_{j}') for j in list(range(m))])
    
t = []
for i in list(range(ncat)):
    t.append([model.NewBoolVar(f't_{i}_{j}') for j in list(range(m))])
    
r = []
for i in list(range(ncla)):
    r.append([model.NewBoolVar(f'r_{i}_{j}') for j in list(range(m))])
    
d = []
for item in items:
    d.append([1 if j in item['binIndices'] else 0 for j in list(range(m))])

for i in list(range(n)):
    model.Add(sum(x[i]) <= 1)
    
for i in list(range(n)):
    for j in list(range(m)):
        model.Add(x[i][j] <= d[i][j])
        model.Add(t[items[i]['t']][j] >= x[i][j])
        model.Add(r[items[i]['r']][j] >= x[i][j])
        
for j in list(range(m)):
    model.Add(sum([x[i][j] * items[i]['w'] for i in list(range(n))]) <= bins[j]['capacity'])
#     model.Add(sum([x[i][j] * items[i]['w'] for i in list(range(n))]) >= bins[j]['minLoad'])
    model.Add(sum([x[i][j] * items[i]['p'] for i in list(range(n))]) <= bins[j]['p'])

for j in list(range(m)):
    model.Add(sum([t[i][j] for i in list(range(ncat))]) <= bins[j]['t'])
    model.Add(sum([r[i][j] for i in list(range(ncla))]) <= bins[j]['r'])

model.Maximize(sum([val for val in [sum(row) for row in x]]))
```

```python
class VarArraySolutionPrinter(cp_model.CpSolverSolutionCallback):
    """Print intermediate solutions."""

    def __init__(self, variables):
        cp_model.CpSolverSolutionCallback.__init__(self)
        self.__variables = variables
        self.__solution_count = 0

    def on_solution_callback(self):
        self.__solution_count += 1
#         for v in self.__variables:
#             print('%s=%i' % (v, self.Value(v)), end=' ')
#         print()

    def solution_count(self):
        return self.__solution_count
    
solver = cp_model.CpSolver()
solution_printer = VarArraySolutionPrinter(x)
status = solver.SolveWithSolutionCallback(model, solution_printer)
print('Number of solutions found: %i' % solution_printer.solution_count())
```

```python
print('x')
for j in list(range(m)):
    s = 0
    p = 0
    print(j)
    for i in list(range(n)):
        if solver.Value(x[i][j]) == 1:
            print(i, items[i]['t'], items[i]['r'], end='\n')
            s += items[i]['w']
            p += items[i]['p']
    print('')
    print(s,p)
    print('')
print('-----')
print('t')
for i in list(range(ncat)):
    for j in list(range(m)):
        print(solver.Value(t[i][j]), end=' ')
    print('')
print('r')
for i in list(range(ncla)):
    for j in list(range(m)):
        print(solver.Value(r[i][j]), end=' ')
    print('')
```
